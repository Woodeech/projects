## Подготовим данные

### Приведем названия всех колонок к нижнему регистру

```python
# применим к колонкам методы str() и lower(), чтобы привести их к нижнему регистру
df_games.columns = df_games.columns.str.lower()
```

```python
# посмотрим первые 5 записей таблицы df_games
display(df_games.head())
```

Теперь все колонки таблицы `df_games` в нижнем регистре.

### Преобразуем данные в нужные типы
Но сначала поменяем значение `tbd` в колонке `user_score` на 0.
TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято

```python
# используем метод replace(), чтобы заменить значение tbd в колонке user_score на 0.
df_games['user_score'] = df_games['user_score'].replace('tbd', np.nan)
```

Поменяем тип данных в колонках:
 - `year_of_release` с float64 на int64
 - `user_score` с object на float64

```python
# используем метод astype(), чтобы поменять тип данных в колонке year_of_release
df_games['year_of_release'] = df_games['year_of_release'].astype('Int64')
# используем метод astype(), чтобы поменять тип данных в колонке user_score
df_games['user_score'] = df_games['user_score'].astype('float64')
```

### Приступим к работе с пропущенными значениями в колонках

Пропуски в колонке `name` составляют `2` шт. менее `1%`. Удалим их. Эти данные большой роли не сыграют.

```python
# используем метод dropna() для удаления пропусков
df_games = df_games.dropna(subset=['name'])
```

```python
# посмотрим количество пропусков в каждой и колонок таблицы df_games
df_games.isnull().sum()
```

Пропуски в колонке `year_of_release` составляют `269` шт. это `1,6%`. Удалим их. Эти данные большой роли не сыграют.

```python
# используем метод dropna() для удаления пропусков
df_games = df_games.dropna(subset=['year_of_release'])
```

```python
# посмотрим долю пропущенных значений от общего количества строк в колонках
display(df_games.isnull().sum()/len(df_games))
```

Остальные пропуски в колонках `other_sales`, `user_score`, `rating` удалять не стоит, т.к. они составляют более `40%` всех данных.

## Исследовательский анализ данных

### Посмотрим, сколько игр выпускалось в разные годы

```python
# создадим сводную таблицу с количеством игр выпушенных в разные годы
year_of_release_pivot_table = df_games.pivot_table(index='year_of_release', values='name', aggfunc='count')
# посмотрим сводную таблицу
display(year_of_release_pivot_table)
```

```python
# Построим график по сводной таблице year_of_release_pivot_table
sns.relplot(data=year_of_release_pivot_table, x='year_of_release', y='name', kind='line').set(title='Количество игр выпущенных в разные годы')
```

## Проверим гипотезы

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезы:

- H0 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
- H1 - Средние пользовательские рейтинги платформ Xbox One и PC различаются

```python
# запишем в переменных xbox_h и pc_h соответствующие данные (актуальные данные)
xone_h = current_period_df[current_period_df['platform']=='XOne']['user_score']
pc_h = current_period_df[current_period_df['platform']=='PC']['user_score']

# посмотрим средний рейтинг пользователя для xbox платформ
display(xone_h.mean().round(2))
# посмотрим средний рейтинг пользователя для pc платформ
display(pc_h.mean().round(2))
```

```python
# посмотрим дисперсию выборок
display(np.var(current_period_df[current_period_df['platform']=='XOne']['user_score']))
display(np.var(current_period_df[current_period_df['platform']=='PC']['user_score']))
```

```python
# используем метод ttest_ind
# В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей
# по взятым из них выборкам:
# Применим метод scipy.stats.ttest_ind (array1, array2, equal_var), 
# где equal_var - параметр, задающий считать ли равными дисперсии выборок. 
# Передаётся как "equal_var = True" или "equal_var = False" (True — считать, False — не считать).

# уровень значимости 5%
alpha = 0.05
# если p-value окажется меньше него - отвергнем гипотезу
# удалим пустые значения с помощью метода dropna()
results = st.ttest_ind(xone_h.dropna(), pc_h.dropna(), equal_var=False)

display('p-значение:', results.pvalue.round(2))


if (results.pvalue < alpha):
    display("Отвергаем нулевую гипотезу")
else:
    display("Не получилось отвергнуть нулевую гипотезу")
```

### Вывод по первой гипотизе
Значение p-value равно чуть меньше `15%`. Не получилось опровергнуть нулевую гипотезу, т.к. среднее рейтинги платформ похожи:
- XOne - `6.52`

Средние пользовательские рейтинги жанров Action и Sports разные

Сформулируем нулевую и альтернативную гипотезы:

- H0 - Средние пользовательские рейтинги жанров Action и Sports одинаковые.
- H1 - Средние пользовательские рейтинги жанров Action и Sports различаются

```python
# запишем в переменных genre_action_h и genre_sports_h соответствующие данные (актуальные данные, все что больше 2009 года)
genre_action_h = current_period_df[current_period_df['genre']=='Action']['user_score']
genre_sports_h = current_period_df[current_period_df['genre']=='Sports']['user_score']

# посмотрим средний рейтинг пользователя для action жанра
display(genre_action_h.mean().round(2))
# посмотрим средний рейтинг пользователя для sport жанра
display(genre_sports_h.mean().round(2))
```

```python
# посмотрим дисперсию выборок
display(np.var(current_period_df[current_period_df['genre']=='Action']['user_score']))
display(np.var(current_period_df[current_period_df['genre']=='Sports']['user_score']))
```

```python
# используем метод ttest_ind
# В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей
# по взятым из них выборкам:
# Применим метод scipy.stats.ttest_ind (array1, array2, equal_var), 
# где equal_var - параметр, задающий считать ли равными дисперсии выборок. 
# Передаётся как "equal_var = True" или "equal_var = False" (True — считать, False — не считать).

# уровень значимости 5%
alpha = 0.05
# если p-value окажется меньше него - отвергнем гипотезу
# удалим пустые значения с помощью метода dropna()
results = st.ttest_ind(genre_action_h.dropna(), genre_sports_h.dropna(), equal_var=False)

display('p-значение:', results.pvalue.round(2))


if (results.pvalue < alpha):
    display("Отвергаем нулевую гипотезу")
else:
    display("Не получилось отвергнуть нулевую гипотезу")
```

### Вывод по второй гипотизе
