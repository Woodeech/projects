## Шаг. Предобработаем данные

Создадим копию таблицы, в которой будем обрабатывать данные

```python
# создадим копию датафрейма и занесем её в переменную df_filtred
df_filtred = df.copy()
```

### Изменим тип данных

Изменим тип данных в колонке `date` с int на datetime

```python
# изменим тип данных в колонке date
df_filtred['date'] = pd.to_datetime(df_filtred['date'], format='%Y%m%d%H')
```

```python
# оставим в колонке дата только дату, а время уберем
df_filtred['date'] = df_filtred['date'].astype('datetime64[D]')
```

### Приведем строки к нижнему регистру

Приведем строки в колонке `product` к нижнему регистру

```python
# приводим строки в колонке "product" к нижнему регистру
df_filtred['product'] = df_filtred['product'].str.lower()
```

### Уберем дубликаты

```python
# удалим дубликаты или отфельтруем датафрейм без них. Сбросим индексы
df_filtred = df_filtred[~df_filtred.duplicated(subset=['customer_id', 'order_id', 'product', 'quantity', 'price'])].reset_index(drop=True)        
```

```python
# проверим наличее заказов совершенных более чем одним покупателем
# сгруппируем данные по заказам и количеству покупателей
duplicated_order_cust = df_filtred.groupby(['order_id']).agg({'customer_id': ['nunique']})['customer_id'].reset_index()

# посчитаем количество покупателей в заказах
display(duplicated_order_cust['nunique'].value_counts())
```

```python
# отфильтруем заказы, в которых больше двух покупателей
dd = duplicated_order_cust[duplicated_order_cust['nunique']>1]
```

```python
# посмотрим сколько всего всего таких заказов
display(dd['order_id'].count())
```

Видно что есть заказы, в которых встречаются сразу несколько покупателей. Такого не может быть. Один заказ - один покупатель. Удалим эти заказы.

```python
# создадим список с номерами заказов, которых больше двух
dd_array = list(dd['order_id'])
```

```python
# удалим эти заказы из списка, которые есть в нашей таблице
df_filtred = df_filtred.query('order_id not in @dd_array')
```

### Добавим необходимые колонки

```python
# добавим столбец total, в котором будем считать количество купленного товара * на цену
df_filtred['total'] = df_filtred.quantity * df_filtred.price
```

### Вывод по предобработке данных
Мы изменили тип данных в колонке `date` и оставили только дату, привели строки в колонке `product` к нижнему регистру. Убрали дубликаты в данных, которых было не видно пока мы не посмотрели на данные бед колонки `date` и удалили заказы, в которых больше чем один покупатель. Узнали минимальную дату - `2018-10-01` и максимальную дату - `2019-10-31` в датафрейме. Добавили колонку `total`, в которой будет общая стоимость покупки.

```python
# напишем функцию, которая будет проходить по всем ключам и их значениям с помощью метода items()
def clustering(adds): 
    for claster, items in dict_products.items():
        for item in items:
            if item in adds:
                return claster
    return 'другое'
```

### Сделаем кластеризацию покупателей

```python
# сгрупируем кластеры покупателей по очищенным данным,
# посчитаем уникальное значение заказов, и средние значения количества, цены и общей стоимости покупки
# сбросим индексы
customers_clasters = df_clear.groupby('customer_id').agg({'order_id' : 'nunique',
                                                           'quantity': 'mean',
                                                           'price':'mean',
                                                           'total':'mean'}).reset_index()

# поменяем названия колонок
customers_clasters.columns = ['customer_id', 'order_id', 'mean_quantity', 'mean_price', 'mean_total']
```

```python
# посмотрим таблицу с кластерами покупателей
display(customers_clasters)
```

```python
# обязательная стандартизация данных перед работой с алгоритмами
sc = StandardScaler()
x_sc = sc.fit_transform(customers_clasters.drop(columns='customer_id'))

# в переменной linked сохранена таблица «связок» между объектами. Её можно визуализировать как дендрограмму
linked = linkage(x_sc, method = 'ward')
```

```python
# укажем размер графика
plt.figure(figsize=(15, 10))

# нарисуем график иерархической кластеризации
dendrogram(linked, orientation='top', no_labels=True)

# заголовок графика
plt.title('Агломеративная иерархическая кластеризация', fontsize = 18)

# уберем технические надписи
plt.show()
```

На основании полученного графика видно, что есть `3` основных кластера. Самый большой из них красный.

### Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры покупателей и посмотрим на средние значения признаков для кластеров

```python
# задаём модель k_means с числом кластеров 3 и фиксируем значение random_state
km = KMeans(n_clusters=3, random_state=0)

# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 2)
labels = km.fit_predict(x_sc)

# сохраняем метки кластера в поле нашего датасета
customers_clasters['clusters'] = labels

# посмотрим первые 5 строк таблицы
display(customers_clasters.head())
```

### Cтатистическая значимость различий среднего чека между кластерами

Для теста будем использовать статестический критерий `Манна-Уитни` основанный на структурном подходе, или непараметрический.

Главное отличие непараметрических методов в том, что они работают с рангами, — номерами значений в упорядоченном ряду, — никак не учитывая сами значения. Поэтому к ним прибегают тогда, когда работа с самими значениями невозможна из-за выбросов, сильно сдвигающих параметрические результаты. В данных есть выбросы, поэтому статистический критерий `Манна-Уитни` здесь подойдет.

Сформулируем гипотезы:
* Нулевая - **Н0**: есть различия в среднем чеке между кластерами.
* Альтернативная - **Н1**: нет различий в среднем чеке между кластерами.

```python
print('Проверяем различия срезнего чека нулевой и первой группы')
print('p-value равен:','{0:.3f}'.format(st.mannwhitneyu(customers_clasters_all[customers_clasters_all['clusters']==0]['total'], customers_clasters_all[customers_clasters_all['clusters']==1]['total'])[1]))
print('Средний чек группы 1:', '{0:.3f}'.format(customers_clasters_all[customers_clasters_all['clusters']==1]['total'].mean()/customers_clasters_all[customers_clasters_all['clusters']==0]['total'].mean()-1))
```

```python
print('Проверяем различия срезнего чека нулевой и второй группы')
print('p-value равен:','{0:.3f}'.format(st.mannwhitneyu(customers_clasters_all[customers_clasters_all['clusters']==0]['total'], customers_clasters_all[customers_clasters_all['clusters']==2]['total'])[1]))
print('Средний чек группы 2:', '{0:.3f}'.format(customers_clasters_all[customers_clasters_all['clusters']==2]['total'].mean()/customers_clasters_all[customers_clasters_all['clusters']==0]['total'].mean()-1))
```

```python
print('Проверяем различия срезнего чека первой и второй группы')
print('p-value равен:','{0:.3f}'.format(st.mannwhitneyu(customers_clasters_all[customers_clasters_all['clusters']==1]['total'], customers_clasters_all[customers_clasters_all['clusters']==2]['total'])[1]))
print('Средний чек группы 2:', '{0:.3f}'.format(customers_clasters_all[customers_clasters_all['clusters']==2]['total'].mean()/customers_clasters_all[customers_clasters_all['clusters']==1]['total'].mean()-1))
```

Нулевая гипотиза не подтвердилась. По всем трем группам статистических различий в среднем чеке нет

### Cтатистическая значимость различий по количеству покупок в среднем на клиента между кластерами есть/нет

Сформулируем гипотезы:
* Нулевая - **Н0**: есть различия между количеством покупок в среднем на клиента между кластерами.
* Альтернативная - **Н1**: нет различий между количеством покупок в среднем на клиента между кластерами.

```python
print('Проверяем различия среднего количества покупок нулевой и первой группы')
print('p-value равен:','{0:.3f}'.format(st.mannwhitneyu(customers_clasters_all[customers_clasters_all['clusters']==0]['quantity'], customers_clasters_all[customers_clasters_all['clusters']==1]['quantity'])[1]))
print('Среднее количество покупок группы 1:', '{0:.3f}'.format(customers_clasters_all[customers_clasters_all['clusters']==1]['quantity'].mean()/customers_clasters_all[customers_clasters_all['clusters']==0]['quantity'].mean()-1))
```

```python
print('Проверяем различия среднего количества покупок нулевой и второй группы')
print('p-value равен:','{0:.3f}'.format(st.mannwhitneyu(customers_clasters_all[customers_clasters_all['clusters']==0]['quantity'], customers_clasters_all[customers_clasters_all['clusters']==2]['quantity'])[1]))
print('Среднее количество покупок группы 2:', '{0:.3f}'.format(customers_clasters_all[customers_clasters_all['clusters']==2]['quantity'].mean()/customers_clasters_all[customers_clasters_all['clusters']==0]['quantity'].mean()-1))
```

```python
print('Проверяем различия среднего количества покупок первой и второй группы')
print('p-value равен:','{0:.3f}'.format(st.mannwhitneyu(customers_clasters_all[customers_clasters_all['clusters']==1]['quantity'], customers_clasters_all[customers_clasters_all['clusters']==2]['quantity'])[1]))
print('Среднее количество покупок группы 2:', '{0:.3f}'.format(customers_clasters_all[customers_clasters_all['clusters']==2]['quantity'].mean()/customers_clasters_all[customers_clasters_all['clusters']==1]['quantity'].mean()-1))
```

Между нулевой и второй группы кластеров есть различия в среднем количестве покупок, а у остальных кластеров статистических различий не выявлено.
