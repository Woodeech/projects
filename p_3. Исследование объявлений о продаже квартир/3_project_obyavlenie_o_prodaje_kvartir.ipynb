Возможно при выгрузке данных произошка техническая ошибка.

```python
# Посмотрим 10 строк с пропущенными значениями с столбце "ceiling_height"
display(df[df['ceiling_height'].isna()].head(10))
```

```python
# Используем метод "fillna()" для заполнения пропусков медианой в связке с методом "transform()". Он создаёт новый столбец,
# который содержит столько же строк, сколько и изначальный,
# но только вместо оригинальных значений в каждой строке содержится медиана посчитанная по заданной группе.
# После этого вместо пропусков в оригинальном столбце мы подставляем значения из созданного столбца,
# то есть медианы в зависимости от группы
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('total_images')['ceiling_height'].transform('median'))
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('days_exposition')['ceiling_height'].transform('median')) 
```

```python
# Посмотрим сумму пропусков в столбце "ceiling_height"
display(df['ceiling_height'].isna().sum())
```

## Расчёты и добавление результатов в таблицу

Расчитаем цену квадратного метра и запишем её в новый столбец.

```python
# Создадим новый столбец "price_per_square_meter". В нём будет расчет цены за квадратный метр
df['price_per_square_meter'] = df['last_price'] / df['total_area']
```

```python
# Изменим тип данных в столбце "price_per_square_meter" на "int"
df['price_per_square_meter'] = df['price_per_square_meter'].astype('int')
```

Создадим столбцы с денем недели, месяцем и год публикации объявления.

```python
# Создадим новый столбец "weekday". Используем атрибут "dt" в связке с "weekday", который выведет день недели
df['weekday'] = df['first_day_exposition'].dt.weekday
```

```python
# Создадим новый столбец "month". Используем атрибут "dt" в связке с "month", который выведет месяц
df['month'] = df['first_day_exposition'].dt.month
```

```python
# Создадим новый столбец "year". Используем атрибут "dt" в связке с "year", который выведет год
df['year'] = df['first_day_exposition'].dt.year 
```

```python
# Посмотрим первые пять строк датафрейма
df.head() 
```

Сделаем категории этажей квартиры с помощью функции. Варианты — первый, последний, другой.

```python
# Напишем фукцию, которая будт проверять, если этаж = 1, то вернет "первый",
# если этаж = всего этажей вернет "последний", во всех остальных случаях вернет "другой"
def floor_category(row): 
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
df['floor_category'] = df.apply(floor_category, axis = 1)
```

```python
# Посмотрим первые двадцать строк датафрейма со столбцами "floor", "floors_total", "floor_category"
df[['floor', 'floors_total', 'floor_category']].head(20) 
```

Сделаем столбец отношения жилой и общей площади, а также отношение площади кухни к общей.

```python
# Создаем новый столбец living area ratio
df['living area ratio'] = df['living_area'] / df['total_area'] 
```

```python
# Создаем новый столбец kitchen_area_ratio
df['kitchen_area_ratio'] = df['kitchen_area'] / df['total_area'] 
```

```python
# Посмотрим первые пять строк датафрейма
df.head() 
```

- категориями этажей;
- отношения жилой и общей площади;
